```python
#Task 2: Handling Missing Values in Project Cost (Running Average Method)

import pandas as pd  # Import pandas
import numpy as np   # Import numpy

project_df = pd.read_csv("Project.csv")  # Load Project dataset

print(project_df)
print(project_df.isnull().sum())         # Count of null values
```

```python
cost_list = project_df["Cost"].tolist()        # Convert Cost column to list

running_sum = 0                                
count = 0                                      

for i in range(len(cost_list)):                # Loop through all cost values
    if pd.isna(cost_list[i]):                  
        cost_list[i] = running_sum / count     # Replace NaN with running average
    else:
        running_sum += cost_list[i]            # Add valid value to running sum
        count += 1                             # Increase valid value count

project_df["Cost"] = cost_list                 # Replace Cost column with updated values
```

```python
project_df
```

```python
project_df["Cost"] = project_df["Cost"].astype(int)  #scientific notation to int conversion
```

```python
#Task 3

employee_df = pd.read_csv("Employee.csv")

# Split Name column into First Name and LastName
employee_df[["First Name", "LastName"]] = employee_df["Name"].str.split(" ", expand=True)

# Drop the old Name column
employee_df.drop("Name", axis=1, inplace=True)

print(employee_df)
```

```python
# Task 4

import pandas as pd

employee_df = pd.read_csv("Employee.csv")
seniority_df = pd.read_csv("Seniority.csv")
project_df = pd.read_csv("Project.csv")

# Merging Employee and Seniority with ID column
final_df = pd.merge(employee_df, seniority_df, on="ID", how="inner")

# Merging the result with Project on ID column
final_df = pd.merge(final_df, project_df, on="ID", how="inner")
```

```python
final_df
```

```python
final_df.to_csv("Final.csv", index=False)  # Saving updates in csv file
```

```python
# Task 5

import numpy as np

final_df["Bonus"] = np.where(final_df["Status"] == "Finished", final_df["Cost"] * 0.05, 0)  # Assign 5% of Cost as Bonus
```

```python
final_df["Bonus"] = final_df["Bonus"].astype(int)  # Convert Bonus values to integer
print(final_df)
```

```python
final_df.to_csv("Final.csv", index=False)  #Saving updates in csv file
```

```python
# Task 6

final_df.loc[final_df["Status"] == "Failed", "Designation Level"] -= 1  # Reduce designation level by 1
```

```python
final_df
```

```python
final_df = final_df[final_df["Designation Level"] <= 4]  # Remove employees with Des Level greater than 4
```

```python
final_df
```

```python
final_df.to_csv("Final.csv", index=False) #Saving updates in csv file
```

```python
# Task 7

import numpy as np

# Add Mr. / Mrs. based on Gender
final_df["First Name"] = np.where(final_df["Gender"] == "M", "Mr. " + final_df["First Name"],"Mrs. " + final_df["First Name"])

# Drop Gender column
final_df.drop("Gender", axis=1, inplace=True)
print(final_df)
```

```python
# Task 8

for i in range(len(final_df)):                     # Loop through each row
    if final_df.loc[i, "Age"] > 29:                # Check if employee age is greater than 29
        final_df.loc[i, "Designation Level"] += 1  # Increase designation by 1
```

```python
final_df
```

```python
# Task 9

TotalProjCost = final_df.groupby(["ID", "First Name"])["Cost"].sum().reset_index() # Group by ID / First Name and calculate sum of Cost per employee

TotalProjCost.rename(columns={"Cost": "Total cost"}, inplace=True)                 # Rename to Total cost
print(TotalProjCost)
```

```python
TotalProjCost.to_csv("TotalProjCost.csv", index=False)   #Saving updates in csv file
```
